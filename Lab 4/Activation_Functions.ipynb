# Activation Functions in Deep Learning


```python
import numpy as np
import matplotlib.pyplot as plt
```

## 1. Sigmoid

#### Define the sigmoid activation function and its derivative

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))

def sigmoid_derivative(z):
    return sigmoid(z) * (1 - sigmoid(z))
```

#### Define the number of input values
Generate 400 synthetic values between -10 and 10. The gradient would be calculated for each of these values.

```python
# Generate a range of input values
z = np.linspace(-10, 10, 400)
```

#### Compute derivative for Sigmoid activations 

```python
sigmoid_grad = sigmoid_derivative(z)
```

```python
plt.figure(figsize=(6,4))
plt.plot(z, sigmoid(z), label='Sigmoid Activation', color='b')
plt.plot(z, sigmoid_grad, label="Sigmoid Derivative", color='r', linestyle='--')
plt.title('Sigmoid Activation & Gradient')
plt.xlabel('Input Value (z)')
plt.ylabel('Activation / Gradient')
plt.legend()

plt.show()
```

## 2. Softmax

Softmax converts a vector of logits into a probability distribution.


$\text{Softmax}(z_i)=\frac{e^{z_i}}{\sum_j e^{z_j}}$


Properties:
- Outputs are between 0 and 1.
- Probabilities sum to 1.
- Used for multi-class classification.


```python
def softmax(x):
    x=x-np.max(x)   # numerical stability
    ex=np.exp(x)
    return ex/np.sum(ex)

logits=np.array([3.0,1.0,0.2])

print("Logits:",logits)

prob=softmax(logits)

print("Softmax probabilities:",prob)
print("Sum:",prob.sum())
```

```python
classes=['Class A','Class B','Class C']

plt.figure(figsize=(6,4))
plt.bar(classes,prob)
plt.ylim(0,1)
plt.ylabel("Probability")
plt.title("Softmax Output")

for i,p in enumerate(prob):
    plt.text(i,p+0.02,f"{p:.3f}",ha='center')

plt.show()
```

## 3. Hyperbolic Tangent (tanh)

Let's plot and compare the output of tanh function with the Sigmoid function for 100 synthetic values between -10 to 10

```python
def tanh(z):
    return np.tanh(z)

def tanh_derivative(z):
    return 1 - np.tanh(z)**2

plt.figure(figsize=(12, 6))

# Plot tanh and its derivative
plt.subplot(1, 2, 1)
plt.plot(z, tanh(z), label='Tanh Activation', color='b')
plt.plot(z, tanh_derivative(z), label="Tanh Derivative", color='r', linestyle='--')
plt.title('Tanh Activation & Gradient')
plt.xlabel('Input Value (z)')
plt.ylabel('Activation / Gradient')
plt.grid(True)
plt.legend()

# Plot Sigmoid and its derivative
plt.subplot(1, 2, 2)
plt.plot(z, sigmoid(z), label='Sigmoid Activation', color='b')
plt.plot(z, sigmoid_grad, label="Sigmoid Derivative", color='r', linestyle='--')
plt.title('Sigmoid Activation & Gradient')
plt.xlabel('Input Value (z)')
plt.ylabel('Activation / Gradient')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```

## 4. ReLU

#### Define the ReLU activation function and its derivative

```python
def relu(z):
    return np.maximum(0, z)

def relu_derivative(z):
    return np.where(z > 0, 1, 0)
```
