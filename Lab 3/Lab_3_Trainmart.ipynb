```python
# Importing the libraries
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
# Importing the dataset
df = pd.read_csv('Train_mart.csv')
```

```python
# 1. Checking the shape of the dataset
print(df.shape)
```

```python
# 2. Distribution of the target variable
import matplotlib.pyplot as plt

plt.hist(df['Item_Outlet_Sales'], bins=30)
plt.xlabel('Target')
plt.ylabel('Frequency')
plt.title('Distribution of Target Variable')
plt.show()
```

```python
# 3. Frequency of outlet size
plt.figure(figsize=(8, 6))
plt.bar(df['Outlet_Size'].value_counts().index, df['Outlet_Size'].value_counts().values)
plt.xlabel('Outlet Size')
plt.ylabel('Frequency')
plt.title('Frequency of Outlet Size')
```

```python
# 4. Correlation between numerical features using seaborn
numerical_features = df.select_dtypes(include=[np.number])
correlation_matrix = numerical_features.corr()


plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Matrix')
plt.show()
```

```python
# 5. Checking for missing values
missing_values = df.isnull().sum()
print("Missing values in each column:\n", missing_values)
```
