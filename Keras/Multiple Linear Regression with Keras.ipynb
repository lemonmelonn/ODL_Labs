# Multiple Linear Regression with Keras


```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from keras.models import Sequential
from keras.layers import Dense, Input
```

```python
house_data = pd.read_csv("kc_house_data.csv")

features = house_data[['sqft_living','grade','sqft_above','sqft_living15']]
target = house_data['price']

display(features.head())
display(target.head())
```

```python
scaler = StandardScaler()
features_norm = scaler.fit_transform(features)

y_scaler = StandardScaler()

target_norm = y_scaler.fit_transform(target.values.reshape(-1,1))

X_train, X_test, y_train, y_test = train_test_split(features_norm,target_norm, test_size=0.2, random_state=42)
```

## Build Model

```python
model = Sequential([Input(shape=(4,)),Dense(1)])

model.summary()
```

```python
model.compile(optimizer='adam',loss='mse',metrics=['mae'])
```

```python
history = model.fit(X_train,y_train,validation_split=0.2,epochs=50,verbose=1)
```

```python
loss, mae = model.evaluate(X_test, y_test, verbose=0)

print("Test MSE:", loss)
print("Test MAE:", mae)
```

```python
new_house = pd.DataFrame([[2500, 8, 2200, 2400]],columns=['sqft_living', 'grade', 'sqft_above', 'sqft_living15'])

new_house_norm = scaler.transform(new_house)
pred_norm = model.predict(new_house_norm)
prediction = y_scaler.inverse_transform(pred_norm)

print(prediction)
```
